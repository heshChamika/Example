```python
from sklearn.datasets import load_wine
w_data = load_wine()
```

```python
w_data.data
```

```python
import pandas as pd
wine_df = pd.DataFrame(w_data.data,columns = w_data.feature_names)
wine_df["target"] = w_data.target
wine_df.head()
```

```python
wine_df.info()
```

```python
wine_df.describe()
```

```python
wine_df.tail()
```

**Data preprocessing**

```python
from sklearn.preprocessing import StandardScaler
X = wine_df[ w_data.feature_names].copy()
y = wine_df["target"].copy()
```

```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X.values)
```

```python
print(X_scaled[0])
```

**model training**

```python
from os import X_OK
from sklearn.model_selection import train_test_split

X_train_scaled,X_test_scaled,y_train,y_test = train_test_split(X_scaled,y,train_size = 0.7,random_state = 25)
print(f"Train size:{round(len(X_train_scaled)/len(X)*100)}% \n\
Test size:{round(len(X_test_scaled)/len(X) *100)}%")
```
